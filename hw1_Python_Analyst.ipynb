## _Условие 2:_
*На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе.*

```python
def fruits_total(**kwargs):
    total = 0
    for key, value in kwargs.items():
        print(f'{key}= {value}')
        total += value
    return total
```

```python
fruits = {'Яблоко': 5, 'Мандарины': 7, 'Слива': 8}

total_f = fruits_total(**fruits)
print(f'{"*" * 10}\nОбщее кол-во фруктов на складе: {total_f}')
```

```python
fruits = {'Яблоко': 5, 'Мандарины': 7, 'Слива': 8, 'Груша': 12, 'Киви': 2, 'Персик': 0}

total_f = fruits_total(**fruits)
print(f'{"*" * 10}\nОбщее кол-во фруктов на складе: {total_f}')
```

```python
''' Ф-ция принимает номер месяца.
    Ф-ция передает назад список индексов из списка list_dates, в которых встречается переданный месяц.
    
    Номера месяцев:
        - Январь: 1
        - Февраль: 2
        ...
        - Ноябрь: 11
        - Декабрь: 12
'''
def list_pos_month(n: int) -> list:
    print(f'>>>Запрос списка индексов на {list_month[n]}.')
    
    # Лямбда-ф-ция, которая возращает число месяца в запрашивемой строке
    month = lambda m: int(list_dates[m].split('-')[1])
    
    # Используем list comprehensions для создания списка индексов из list_dates, 
    # строки которых содержат запрашиваемый месяц.
    return [i for i in range(len(list_dates)) if month(i) == n]
```

```python
# Запрашиваем список индексов из list_dates, строки которых содержат запрашиваемый месяц ноябрь (11)
list_find_dates = list_pos_month(11)
print(f'Позиции индексов Ноября: {list_find_dates}')
```

```python
# Лямбда-ф-ция, которая из списка list_purchases возращает число по индексу i
purchas = lambda i: list_purchases[i]

# Список с ноябрьским суммами 
purchas_november = [purchas(i) for i in list_find_dates]

print(f'{"*" * 10}\nВыручка компании в ноябре: {sum(purchas_november)}.')
```
